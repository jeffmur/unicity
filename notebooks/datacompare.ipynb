```python
import pandas as pd 
import numpy as np 
import datacompy as dcomp 
from lib import geoSanitize as geo, preprocess as pre, unique
```

```python
## Compare Two Users to find high similarity
## Try to get quick metric for efficent exhaustive search.
# Just raw data compare.
zero = geo.filterUserMonthRange('000', '2008-10', '2008-11')
one = geo.filterUserMonthRange('001', '2008-10', '2008-11')
```

```python
## Only keep 2 decimals of lon/lat location
zeroUserMonth = unique.locationPrecision(zero, 2)
oneUserMonth = unique.locationPrecision(one, 2)


# Create location ids for each unique location
zeroUserMonth['loc_id'] = unique.generateLocationID(zeroUserMonth)
oneUserMonth['loc_id'] = unique.generateLocationID(oneUserMonth)
```

```python
compare = dcomp.Compare( 
    zeroUserMonth[['loc_id']], oneUserMonth[['loc_id']], 
    df1_name='000', df2_name='001',
    on_index=True
)

print(compare.report())
```

```python
together = pd.concat([oneUserMonth, zeroUserMonth])
unique.inTheCrowd(5, together, together.groupby("UID"))
```

```python
## Now with a bounding box 
bb = pre.fetchGeoLocation("Beijing, China")
print(bb)
zeroUserMonth = pre.dropOutlyingData(zeroUserMonth, bb)
oneUserMonth = pre.dropOutlyingData(oneUserMonth, bb)
```

```python
compare = dcomp.Compare( 
    zeroUserMonth[['loc_id']], oneUserMonth[['loc_id']], 
    df1_name='000', df2_name='001',
    on_index=True
)

print(compare.report())
```

```python
together = pd.concat([oneUserMonth, zeroUserMonth])
unique.inTheCrowd(5, together, together.groupby("UID"))
```
