```python
import pandas as pd 
import numpy as np 
import datacompy as dcomp 
from lib import geoSanitize as geo, preprocess as pre, unique
```

```python
userSet = ['000','001','002']
df = geo.multiUserDF(userSet)
```

```python
print(f"df with users: {userSet}")
print(f"Longitude [{min(df.Longitude)},{max(df.Longitude)}]")
print(f"Latitude [{min(df.Latitude)},{max(df.Latitude)}]")
boundingBox = pre.fetchGeoLocation('Bejing, China')
print(f"Bounding Box: {boundingBox}")
```

```python
beijing_df = pre.dropOutlyingData(df, boundingBox)
beijing_df = unique.locationPrecision(beijing_df, 2)
beijing_df['loc_id'] = unique.generateLocationID(beijing_df)
```

```python
u = unique.inTheCrowd(5, beijing_df, beijing_df.groupby('UID'))
userSimilarity = 100 - (u*100)
print(f'{round(userSimilarity)}% similarity')
```

```python
# Very Granular Location... at the University ...
# Or atleast records where all users have passed it :)
df = geo.multiUserDF(geo.getAllUID())
boundingBox = pre.fetchGeoLocation("University of Science and Technology Beijing, China")
print(f"Bounding Box: {boundingBox}")
```

```python
university = pre.dropOutlyingData(df, boundingBox)
university = unique.locationPrecision(university, 3)
university['loc_id'] = unique.generateLocationID(university)
print(university)
```

```python
u = unique.inTheCrowd(5, university, university.groupby('UID'))
userSimilarity = 100 - (u*100)
print(f'{round(userSimilarity)}% similarity')
```
